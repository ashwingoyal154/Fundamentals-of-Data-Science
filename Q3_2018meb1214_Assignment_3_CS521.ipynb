```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from sklearn.datasets import load_iris
data=load_iris()

X=data.data
y=data.target
a=data.target_names
df=pd.DataFrame(X,columns=data.feature_names)
df['Class']=y
df
```

```
#from sklearn.preprocessing import MinMaxScaler
#scaler = MinMaxScaler()
#X=scaler.fit_transform(X)
```

```
df=df.drop(columns=["sepal width (cm)","sepal length (cm)"])
df
```

```
X=df[['petal length (cm)','petal width (cm)']]
#X=df[['sepal length (cm)','sepal width (cm)']]
X=X.to_numpy()

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

from sklearn import svm
from sklearn.svm import SVC

model=SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred=model.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

clf=svm.SVC(kernel='linear')
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y,s=40)
plt.title("Using SVM on all 3 classes and 2 features")
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```
X=df[['petal length (cm)','petal width (cm)']]
X=X.to_numpy()

X=X[50:150,:]
y=y[50:150]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

from sklearn import svm
from sklearn.svm import SVC

model=SVC(kernel='linear')
model.fit(X_train, y_train)
y_pred=model.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))

clf=svm.SVC(kernel='linear')
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y,s=40)

ax = plt.gca()

X1 = np.linspace(ax.get_xlim()[0], ax.get_xlim()[1], 30)
y1 = np.linspace(ax.get_ylim()[0], ax.get_ylim()[1], 30)
y2, X2 = np.meshgrid(y1, X1)

stacked = np.vstack([X2.ravel(), y2.ravel()]).T #returns a flattened array
z = clf.decision_function(stacked).reshape(X2.shape)

ax.contour(X2, y2, z, colors='black', levels=[-1, 0, 1], linestyles=['--', '-', '--'])
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=110,facecolors='none', edgecolors='black')

plt.title("virginica and virsicolor")
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
