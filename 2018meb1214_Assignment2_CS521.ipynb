```
import numpy as np
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from sklearn.datasets import load_iris
data=load_iris()

X=data.data
y=data.target
a=data.target_names
df=pd.DataFrame(X,columns=data.feature_names)
df['Class']=y
df
```

```
from sklearn import preprocessing
data.data = preprocessing.scale(data.data)
#checking 
df.isnull().sum()
```

```
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split

#X=data.data
#y=data.target
kf = KFold(n_splits=5, shuffle=False)

for train,test in kf.split(X):
  X_train, X_test=X[train], X[test]
  y_train, y_test=y[train], y[test]

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
```

# Prediction

```
from sklearn.linear_model import LogisticRegression
from sklearn import model_selection

lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))

#5 fold cross validation
kfold = model_selection.KFold(n_splits=5)
cv_res = model_selection.cross_val_score(LogisticRegression(), X_train, y_train, cv=kfold, scoring='accuracy')

print(cv_res.mean(), cv_res.std())
```

```
from sklearn import model_selection
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()

gnb.fit(X_train,y_train)
y_pred_gnb=gnb.predict(X_test)

prob_gnb=gnb.predict_proba(X_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred_gnb))
print(confusion_matrix(y_test,y_pred_gnb))

#5 fold cross validation
kfold = model_selection.KFold(n_splits=5)
cv_res = model_selection.cross_val_score(GaussianNB(), X_train, y_train, cv=kfold, scoring='accuracy')

print(cv_res.mean(), cv_res.std())
```

# DATA CLASSIFICATION

```
#K-means

from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(X)
kmeans.labels_
y_pred_kmeans=kmeans.fit_predict(X)

plt.scatter(X[y_pred_kmeans==0,0],X[y_pred_kmeans==0,1],c='red')
plt.scatter(X[y_pred_kmeans==1,0],X[y_pred_kmeans==1,1],c='blue')
plt.scatter(X[y_pred_kmeans==2,0],X[y_pred_kmeans==2,1],c='yellow')

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y,y_pred_kmeans))
print(confusion_matrix(y,y_pred_kmeans))
```

```
#from the heatmap plotted above, clearly petal width and petal length are more correlated with the class(0,1,2) and are more useful features
#hence I take the subset of X here and use kmeans again to classify
x=X[:,[2,3]]

#K-means
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3)
kmeans.fit(x)
kmeans.labels_
y_pred_kmeans=kmeans.fit_predict(x)

plt.scatter(x[y_pred_kmeans==0,0],x[y_pred_kmeans==0,1],c='red')
plt.scatter(x[y_pred_kmeans==1,0],x[y_pred_kmeans==1,1],c='blue')
plt.scatter(x[y_pred_kmeans==2,0],x[y_pred_kmeans==2,1],c='yellow')

#from sklearn.metrics import classification_report,confusion_matrix
#print(classification_report(y,y_pred_kmeans))
#print(confusion_matrix(y,y_pred_kmeans))

#after seeing the plot below, it is obvious that the misclassifications were due to the wrongness of the results predicted through sepal width and sepal length,
# which are not very good predictors (this can be seen from their low correlation in the heatmap)
```
