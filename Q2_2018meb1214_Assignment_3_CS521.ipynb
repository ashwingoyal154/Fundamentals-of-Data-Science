```
import numpy as np
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```
from sklearn.datasets import load_iris
data=load_iris()

X=data.data
y=data.target
a=data.target_names
df=pd.DataFrame(X,columns=data.feature_names)
df['Class']=y
df
```

```
X
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X=scaler.fit_transform(X)
```

```
#PCA
from sklearn.decomposition import PCA

comp=100
pca=PCA(n_components=2)
pca.fit(X)

X_pca=pca.transform(X)
#X_test_pca=pca.transform(X_test)
#X_train_pca.shape
X_pca.shape
```

```
a1=[]
a2=[]
a3=[]
a4=[]
colors=np.array(['r','b','g'])
for i in range(0,150):
    if df['sepal length (cm)'][i]> df['sepal length (cm)'].mean():
        a1.append(0)
    else:
         a1.append(1)
    if df['sepal width (cm)'][i]> df['sepal width (cm)'].mean():
        a2.append(0)
    else:
         a2.append(1)         
    if df['petal length (cm)'][i]> df['petal length (cm)'].mean():
        a3.append(0)
    else:
         a3.append(1)
    if df['petal width (cm)'][i]> df['petal width (cm)'].mean():
        a4.append(0)
    else:
        a4.append(1)
```

```
def irisplot(x, y, title, xlabel="1st eigenvector", ylabel="2nd eigenvector"):
    plt.scatter(x, y,c=colors[a1],s=55 ) #c=df['sepal length (cm)'],
    plt.title(title, fontsize='15')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.text(0,0.9,'Red: High values (Above average)')
    plt.text(0,0.8,'Blue: Low values (Below average)')

irisplot(x=X_pca[:,0],y=X_pca[:,1],title=' PCA- Characterization through sepal length(2 dimensions)')
```

```
fig = plt.figure(figsize = (8,8))

setosa_versicolor = data.data[0:100,0:3]
setosa_versicolor_tgts = data.target[0:100]
versicolor_virg = data.data[50:150,0:3]
versicolor_virg_tgts = data.target[50:150]

setosa_virginica = np.concatenate((data.data[0:50,0:3], data.data[100:150,0:3]), axis=0)
setosa_virginica_tgts = np.concatenate((data.target[0:50], data.target[100:150]), axis=0)
s1=["setosa","versicolor"]
s2=["versicolor","virginica"]
s3=["setosa","virginica"]

def separation(X_lda, targets, id_1, id_2):
    mean_class_1 = X_lda[targets==id_1,0]
    mean_class_2 = X_lda[targets==id_2,0]
    result_mean = (np.mean(mean_class_1) + np.mean(mean_class_1))/2
    return result_mean.round(decimals=15)

def scatterPlot(X_lda,targets,id_1, id_2,text):
    for color, i, a1 in zip(['red', 'blue'], [id_1, id_2], a):
        plt.scatter(X_lda[targets==i,0], np.zeros([50]), color=color, label=a1)
        plt.title("Scatter plot for " + text + " classification")
    #plt.legend()
    plt.plot([0,0],[-0.04,0.04])
    plt.show()
    #plt.plot(str(s),0,'x')

lda_result = (lda.fit(setosa_versicolor, setosa_versicolor_tgts).transform(setosa_versicolor))

scatterPlot(lda_result,setosa_versicolor_tgts,0,1,"setosa and versicolor")
s = separation(lda.fit(setosa_versicolor, setosa_versicolor_tgts).transform(setosa_versicolor),setosa_versicolor_tgts,0,1)  
print("Threshold:" + str(s))
plt.axvline(str(s),0,1, label='Threshold line')
plt.legend()
```

```
fig = plt.figure(figsize = (8,8))
lda_result = (lda.fit(setosa_virginica,setosa_virginica_tgts).transform(setosa_virginica))

scatterPlot(lda_result,setosa_virginica_tgts,0,2,"versicolor and virginica")
s = separation(lda.fit(setosa_virginica,setosa_virginica_tgts).transform(setosa_virginica),setosa_virginica_tgts,0,2)
print("Threshold:" + str(s))
plt.axvline(str(s),0,1, label='Threshold line')
plt.legend()
```

```
fig = plt.figure(figsize = (8,8))

lda_result = (lda.fit(versicolor_virg,versicolor_virg_tgts).transform(versicolor_virg))

scatterPlot(lda_result,versicolor_virg_tgts,1,2,"setosa and virginica")
s = separation(lda.fit(versicolor_virg,versicolor_virg_tgts).transform(versicolor_virg),versicolor_virg_tgts,1,2)
print("Threshold:" + str(s))
plt.axvline(str(s),0,1, label='Threshold line')
plt.legend()
```

```
from mpl_toolkits.mplot3d import Axes3D
def irisplot3D(x, y,z, title, xlabel="Dimension 1", ylabel="Dimension 2",zlabel="Dimension 3"):
    fig = plt.figure(1, figsize=(8, 8))
    axes = Axes3D(fig, elev=-150)
    axes.scatter(x, y,z,c=y,s=45 ) 
    axes.set_title(title, fontsize='25')
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.set_zlabel(zlabel)
```

```
#tsne 3D with metric= Euclidean
from sklearn.manifold import TSNE

tsne = TSNE(n_components=3, n_iter=2000, metric='euclidean')
X_tsne = tsne.fit_transform(X)

irisplot3D(x=X_tsne[:,0],y=X_tsne[:,1],z=X_tsne[:,2],title='(metric=euclidean)Iris Dataset in 3 dimensions through t-SNE ')
```
